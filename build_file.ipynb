## 2. Currency rounding to cent

### Let's start with removing the "$" and ","  for the wholesale column

```python
df['wholesale ($)'] = df['wholesale ($)'].astype(str)
df['wholesale ($)']
```

```python
def normalize_number(str):
    if '$' in str:
        str = str.replace('$', '')
    if ',' in str:
        str = str.replace(',', '')
    return str

    
    
```

```python
for index, row in df.iterrows():
    df.at[index, 'wholesale ($)'] = normalize_number(row['wholesale ($)'])
```

### Round to two decimal places

```python
df['wholesale ($)'] = df['wholesale ($)'].astype(float).map('{:.2f}'.format)
```

### Map column

```python
df['map ($)'] = df['map ($)'].astype(str)
```

```python
for index, row in df.iterrows():
    df.at[index, 'map ($)'] = normalize_number(row['map ($)'])
```

```python
df['map ($)'] = df['map ($)'].astype(float).map('{:.2f}'.format)
```

### Now the MSRP column

```python
df['map ($)'] = df['map ($)'].astype(float).map('{:.2f}'.format)
```

### replacement glass price ($) column

```python
df['replacement glass price ($)'] = np.where(df['replacement glass price ($)'].notna(), normalize_number(df['replacement glass price ($)'].astype(str)), df['replacement glass price ($)'])
```

```python
def normalize_value(row, column):
    if not pd.isna(row[column]):
        return normalize_number(row[column])
    else:
        return row[column]
```

```python
df['replacement glass price ($)'] = df.apply(normalize_value, axis=1, column='replacement glass price ($)')
```
