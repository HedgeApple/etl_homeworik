### Transformations

```python
# Define a function to format "ean13" column values
def format_number(number):
    number_str = str(number)  # Convert the number to a string
    return '0'+ number_str[:2] + '-' + number_str[2:11] + '-' + number_str[11:]
```

```python
# Apply the function to "ean13" columns
formatted['ean13'] = formatted['ean13'].apply(format_number)
```

```python
# In the above formatting we have a drawback that it returns value with a decimal so to remove the decimal
# Define a function to format the decimal
def format_number(number):
    return number.split(".")[0]
```

```python
# Apply the function to "ean13" columns
formatted['ean13'] = formatted['ean13'].apply(format_number)
```

```python
# Define a function to format the currency values
def format_currency(value):
    value = value.replace('$', '')  # Remove dollar sign
    value = value.replace(',', '')  # Remove commas
    value = float(value)  # Convert to float
    rounded_value = round(value, 2)  # Round to two decimal places
    formatted_value = "${:,.2f}".format(rounded_value)  # Add dollar sign and format as currency
    return formatted_value
```

```python
# Apply the function to the desired columns
formatted['cost_price'] = formatted['cost_price'].astype(str).apply(format_currency)
formatted['min_price'] = formatted['min_price'].astype(str).apply(format_currency)
```
